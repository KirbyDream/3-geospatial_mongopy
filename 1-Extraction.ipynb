# Selecting companies :
In this notebook, we're going to select companies with criterias relevant to our search, and turn the result we get into a clean dataFrame.

## Starting from the "companies" database in Mongo, and importing only the companies :
 1-that do anything related to gaming design..  
 2- that have total money raised greater than 10M$.
 
 
 
 ( instead of 1M$, to have a smaller sample size)

```python
data_base.distinct("total_money_raised")
```

```python
condition1 = {"total_money_raised" : {"$regex": "\d{2,}\.?\d?M|\d{1,}\.?\d?B"}}
condition2 = {"tag_list": {"$regex": ".*design.|.tech.|.software.|.fashion.|.game."}}
condition3 = {"category_code": "web"}
condition4= {"offices":{"$ne":[]}}
projection = {"_id": 0, "name":1, "offices.city":1, "offices.latitude":1, "offices.longitude":1,"offices.country_code":1 }
query = {"$and": [condition1, condition4, {"$or": [condition2, condition3]}]}
result = list(data_base.find(query, projection))
              
              
                            
```

```python
len(result)
```

```python
df = pd.DataFrame(result)
df
```

```python
df.offices[0]
```

```python
df = df.explode("offices")
df
```

```python
df = df.reset_index()
df
```

```python
cities = []
latitude = []
longitude = []
country = []

for index, row in df.iterrows():
    try:
        cities.append(row["offices"]["city"])
        latitude.append(row["offices"]["latitude"])
        longitude.append(row["offices"]["longitude"])
        country.append(row["offices"]["country_code"])                
    except IndexError :
        cities.append(None)
        latitude.append(None)
        longitude.append(None)
        country_code.append(None)

df["city"] = cities
df["latitude"]= latitude
df["longitude"]= longitude
df["country_code"]= country
```

```python
#drop offices
new_df = df.drop(["offices"], axis = 1)
new_df
```
